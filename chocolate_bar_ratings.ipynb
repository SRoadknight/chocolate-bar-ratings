## 1.2 Supervised learning task identification

The research teams goal is to predict the rating a chocolate bar (with a foucs on plain dark chocolates) would receive on the Flavors of Cocao website based on the dataset we have. The target variable to be used is the rating that the chocolates have been given. This will form the ground truth, ground truth are the values given to the target labels in the dataset as these are what the supervised learning algorithm will use to train the model (c3.ai, n.d.a) , the ground truth was established by the reviewer of the chocolate bar by assessing the flavour, texture and after melt of the chocolate bar objectively, as well as their overall opinion. Ratings are on a scale of 1 - 5 with the ability to have a fraction of .00, .25, .50, and .75 to show the variance of dark chocolate bars in the same range, such as 1.25 and 1.75.

<br>

The target variable rating has an ordinal statistical data type which can be treated as a continuous quantity, therefore the research team have chosen to treat our problem as a regression task. Regression algorithms (Brownlee, 2017a)  are a family of supervised machine learning methods that are inteded to predict values which are continuous.

<br>

This produced model would benefit chocolate manufacturers when deciding what type of cocoa bean(s) to use, where they should source them from and what cocoa percentage the chocolate bar should be. Allowing them to make more informed decision when developing a new chocolate bar, and specifically a plain dark chocolate bar.

## 2.2 Splitting of the dataset

#### Why do we split the dataset?

The reason for splitting the dataset into training and testing  sets is to allow us to have a portion of data that can be treated as “seen” data, the training set and another that can be thought of as “unseen” data the testing set. The training set represents data that we have collected, allowing us to perform exploratory data analysis (EDA) on it, as well as using it train our machine learning (ML) model, whereas the testing set represents data that is yet to be collected, but could be possible data our model encounters to predict on in the future. This prevents data leakage (Brownlee, 2016a) , having access to data that shouldn't be available or "seen" and using it to gain knowledge and influence the modelling process. We further enhance this with a validation set, which is an intermediary, and can initially be thought of as "unseen" data during the EDA process.  

<br>

All three of the training, validation and testing sets are samples of the original source data, and they should all be representitive sample of the source data. Once we have carried out our splits we can do some basic analysis without leaking any patterns in the data to help persuade ourselves that this is the case.

#### Temporally splitting the data

```python
# Returns the features and target of the DataFrame as X and Y respectively
def X_Y_splitter(df, target_column):
  X = df.drop(target_column, axis=1)
  Y = df[target_column]
  return X, Y

# Get the DataFrame using the boolean index, the DataFrame will consit of only
# the records which are True.
def df_in_range(df, bool_series):
  ranged_df = df[bool_series]
  return ranged_df

# Helper function to split dataframe into train, validate, and test sets
# Taking in as parameters, the DataFrame to be split,
# the column it will be filtered on (review_year)
# the target column (rating)
# and the maximum dates we want for both the train set and validation set
def train_validate_test_split(df, filter_column, target_column, train_date_max=2013, validate_date_max=2014):
  # Setup the conditions for each of the splits and get the boolean index
  upper_range_cond_train = df[filter_column] <= train_date_max
  range_cond_validate = (~upper_range_cond_train) & (df[filter_column] <= validate_date_max)
  lower_range_cond_test = (~upper_range_cond_train) & (~range_cond_validate)
  # Using the boolean index's created to get the df's for train, validate, and test
  df_train = df_in_range(df, upper_range_cond_train)
  df_validate = df_in_range(df, range_cond_validate)
  df_test = df_in_range(df, lower_range_cond_test)
  # Split the train, validate, and test DataFrame into their X features and Y target.
  X_train, Y_train = X_Y_splitter(df_train, target_column)
  X_validate, Y_validate = X_Y_splitter(df_validate, target_column)
  X_test, Y_test = X_Y_splitter(df_test, target_column)
  return df_train, df_validate, df_test, X_train, X_validate, X_test, Y_train,\
   Y_validate, Y_test

# Concat the X and Y values to create the DataFrame of features and target.
def X_Y_concat(X, Y):
  X_y = pd.concat([X, Y], axis=1)
  return X_y
```

```python
# Using the created helper function to split the data into train,
# validate and test
df_train, df_validate, df_test, X_train, X_validate, X_test, Y_train, Y_validate, Y_test = \
  train_validate_test_split(df, "review_year", "rating")
```

Instead of using derivatives of the bean types, we can group the beans into four main types, Criollo, Trinitario, Nacional, and Forastero, along with Blend, for known blends.

<br>

To create the parse_bean_type function, information about the unique bean type values from each set has been used. This means we are using data from the validation and test data, while this allows for some leakage of the data, however, it doesn't expose any patterns or information about the data that would bias future decisions.  

```python
def blend_count(text):
  """
  Helper function for parse_bean_type to determine to help determine if a bean_type
  is a blend or not
  """
  bean_types = ["criollo", "trinitario", "nacional", "forastero"]
  if "blend" in text:
    return 0
  return sum(1 if bean_type in text else 0 for bean_type in bean_types)

def parse_bean_type(text):
  """
  Parses the value and returns the variety of bean.
  """
  text = text.lower()
  count = blend_count(text)

  # After looking at the training, validation, and blend sets, we could conclude
  # if the bean type has the word blend or
  if any(x in text for x in ["blend", "amazon"]):
    return "Blend"
  if "trinitario" in text:
    return "Trinitario" if count == 1 else "Blend"
  if text.startswith("forastero"):
    return "Forastero"
  if text.startswith("nacional"):
    return "Nacional"
  if text.startswith("criollo"):
    return "Criollo"
  return "Unclassified"
```

We will now use the helper function to crete a new DataFrame object that contains the mapped bean_type column. This will be used to classify each bean type into either one of the main classes of cocoa beans, Blend, or Unclassified.

```python
# Use the created function to created a DataFrame with the mapped bean types.
continent_tmp = deepcopy(df_train)
assert(continent_tmp['bean_type'].dtype == 'O')
continent_tmp['bean_type'] = continent_tmp['bean_type'].apply(parse_bean_type)
```

<br>

| Quality | Description |
|---------|-------------|
| Flavour | The flavour makes up the most important part of the rating, considering the diversity, balance, intensity and purity of favlours. Topics discussed caninclude genetics, terroir, post harvest tehcniuqes, processing and storage as they can have an effect on the flavour |

| Aftermelt | The experience once the chocolate has melted, a higher quality chocolate will linger and be enjoyable. This is the last impression made from the chocolate and is reflected so in the rating |
| Overall opinion | This is the reviewers overall opinion and is the subjective part of the review, taking into account the qualities above and bringing them together as a whole. This also allows for summarisation of the chocolate and its impression on the reviewer |

While chocolate artisians aim to express the flavour of the cocoa that is grown there are other inital factors to consider such as the genetics of the cocoa tree, the flavour of the fruit itself, and the microbes involved in fermentation (Bean to Bar World, n.d.a) , as well as human decisions in the process, such as farming techniques, drying methods and fermentation processes. With all this factoring in to the flavour of the chocolate (Bean to Bar World, n.d.b) , origin or bean type alone may not be a great predictor of rating as the influences of flavour can change in the same origin from region to region..

As the bean type variable has ~44% of its values missing, broad bean origin would be favourable, however as we've discovered bean type appears to be a better predictor of rating.

To get an initial look at the correlation I have chosen to use Cramer's V (a.k.a Cramer's Phi), it allows us to measure the correlation of two nominal values, giving a value between 0 and +1, it is based on the Pearson's chi-squared statistic. We will be using the bias corrected version.

```python
def cramers_corrected_stat(confusion_matrix):
    """ calculate Cramers V statistic for categorial-categorial association.
        uses correction from Bergsma and Wicher,
        Journal of the Korean Statistical Society 42 (2013): 323-328
        https://stackoverflow.com/questions/20892799/using-pandas-calculate-cram%C3%A9rs-coefficient-matrix
    """
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.to_numpy().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))
```

With a result of ~0.59 meaning there is a moderate to strong correlation, it will be worth exploring in the moddeling stages whether we could drop either the bean type or the broad bean origin without losing too much information and reducing the performance of the model.

```python
# Uses the assumption that df_train has had the no-break spaces and NaN
# values replaced with "Unspecified"
# This is the case for no missing data, and the bean type "cleaned up"
bean_type_origin_tmp = deepcopy(df_train)
bean_type_origin_tmp = bean_type_origin_tmp[bean_type_origin_tmp['bean_type'] != "Unspecified"]
bean_type_origin_tmp['bean_type'] = bean_type_origin_tmp['bean_type'].apply(parse_bean_type)
confusion_matrix = pd.crosstab(bean_type_origin_tmp['broad_bean_origin'], bean_type_origin_tmp['bean_type'])
cramers_corrected_stat(confusion_matrix)
```

Using the plot below along with the value calculated from Cramer's V, we can see that in some cases the origin is a good indication to which bean type might have been used such as Madagascar, but then there are others such as Venezuela which have a mix. This may be the case of why it is moderate to strong, with some of the origin with the most data points having a mix on bean types.

## 3.1 Identification of chosen supervised learning algorithm



Support Vector Regression (SVR) (Smola and Schölkopf, 2004) is the chosen supervised learning algorithm for the identified regression task.

<br>

SVR uses the same principles as Support Vector Machines (SVM) which is to find the best line of fit (hyperplane) to classify the data points. SVR can be used to predict discrete values. To do this we define the decision boundaries (margins) in which our hyperplane/regression line (line that helps predict the continuous data) lies between which can be thought of as our tolerance for error as it is error insensitive, the data points that fall within the boundaries are known as support vectors and will not be penalised for their error, while data points outside of the boundary will. There are different kernels that can be used to map our lower dimensional data to higher dimensional data, allowing for the hyperplane and decision boundaries to be calculated.


<br>

SVR doesn't scale well to large datasets with the fit time complexity being larger than quadratic with the number of samples, meaning it doesn't scale well to samples in the tens of thousands (Scikit-learn, n.d.a) . Our dataset has less than2000 samples in total, and just over 1000 in the training set that will be used to fit the model.

<br>

SVR is appropriate for our learning task as we are predicting the continuous value that is the rating of the chocolate bar, it supports both linear and non-linear kernels, allowing for experimentation and some further EDA during the modelling process.

## 3.2 Identification of appropriate evaluation techniques



Evaluation metrics are used both in the process of training the model and testing the model, before productionalisation of the model and during model monitoring. They allow us to evaluate the performance of the ML model, and inform our decision on the steps that should be taken. Different evaluation metrics reveal different information about the model, it is worth considering using multiple evaluation metrics where each reveals different information about the model.

The research team chose three evaluation metrics, Mean Absolute Error (MAE), Mean Squared Error (MSE) and R-Squared.



### Mean Absolute Error

Mean Absolute Error (MAE) (Brownlee, 2021a) is an error metric for regression problems with the goal to measure the error of the model. The MAE represents error of all predictons using the same scale the predictions were on. It is calculated by summing up the absolute errors (absolute describing the distance from zero, and error being the actual value - predicted value) and then dividing by the sample size.

<br>
Formula for MAE:
<br><br>
$x$ and $y$ are $D$ dimensional vectors, and $x_i$ denotes the value on the $i$th dimension of $x$.

\begin{align}
\sum_{i=1}^{D}|x_i-y_i|
\end{align}


<br>

This metric was chosen as it can show us in absolute terms the aggregated mean error across all the predictions while being on the same scale as the data which gives it context to our target variable.

### Mean Squared Error

Mean Squared Error (MSE) (Brownlee, 2021b) is an error metric for regression problems similar to MAE, both having the same goal. It is calculate differently to MAE, summing up the square of each error and then dividing by the sample size. As it sums up the square of each error it punishes larger errors more than smaller error, whereas MAE does not give more or less weight to different errors, as it increases linearly with the increase in error.

<br>
Formula for MSE:
<br><br>
$x$ and $y$ are $D$ dimensional vectors, and $x_i$ denotes the value on the $i$th dimension of $x$.
\begin{align}
\sum_{i=1}^{D}(x_i-y_i)^2
\end{align}


<br>

This was chosen as our data has some "extreme" ratings on both the lower and upper end of the scales, we can use MSE to potentially detect these and work towards minimising them.

<br>

Although both MAE, and MSE are error metrics, we are using one for interpetability (MAE), while we are using the other one in the hopes of detecting and reducing any large errors.

### R-Squared

R-squared (R²) (Frost, n.d.) is an evaluation metric for regression problems, R² represents the proportion of variance in the dependant variable that can be explained by the independant variables, which for us means the proportion of variance in rating based on the features we have chosen.

<br>

The output of R² is between 0 and 1, with 0 showing no variance in the rating based on the independant features and 1 shows full explainibility in rating based on our chosen independant features.

<br>

We have chosen this metric as it doesn't represent "error" like MAE and MSE do, it instead tells us how well the data fit the regression model.

### Computational Performance

As well as looking at the predictive performance of a model we may also want to look at the computational peformance of a model. Looking at the memory taken when performing computation as well as the time taken to make a prediction. Depending on the goal of the system and what is being predicted, we may favour computation performance over predictive performance, such as online shopping websites, however for other tasks predictive performance may be held in higher regard than the computational performance such as in medical diagnosis.

We will now use the gained knowledge from our EDA to clean and pre-process the data so that it can be used to train an ML model, the initial pre-processing was agreed upon by myself and the research team.

First we will specify our helper functions, these will allow us to create different models without re-writing the same code for some common tasks that will be carried out.

```python
# fill missing values -- the no-break spaces
def fill_missing(df):
  """
  """
  df = df.apply(lambda x: x.replace('\xa0', np.NaN))
  df = df.fillna("Unspecified")
  return df

# drop desired columns
def drop_columns(df, columns):
  """
  """
  df = df.drop(columns, axis=1)
  return df

# Encode cocoa percent
def encode_cocoa_percent(df):
  """
  """
  cocoa_percent = \
  (df['cocoa_percent'].apply(lambda x: str(x).replace('%', '')).astype(float)) / 100
  return cocoa_percent

def blend_count(text):
  """
  """
  bean_types = ["criollo", "trinitario", "nacional", "forastero"]
  if "blend" in text:
    return 0
  return sum(1 if bean_type in text else 0 for bean_type in bean_types)

# Parse bean variety
def parse_bean_type(text):
  """
  """
  text = text.lower()
  count = blend_count(text)

  if any(x in text for x in ["blend", "amazon"]):
    return "Blend"
  if "trinitario" in text:
    return "Trinitario" if count == 1 else "Blend"
  if text.startswith("forastero"):
    return "Forastero"
  if text.startswith("nacional"):
    return "Nacional"
  if text.startswith("criollo"):
    return "Criollo"
  return "Unclassified"

# Parse company locations to continents (Modified version of Hitesh's so we use a dict rather than looping through lists for each value)
# This is messy and could be cleaned up
def create_country_continent_dict(country_lists, continent_list):
    country_continent_dict = {}
    for country_list, continent in zip(country_lists, continent_list):
        country_continent_dict.update(dict.fromkeys(country_list, continent))
    return country_continent_dict

europe = ['France','Switzerland','Netherlands','Spain','Italy','U.K.','Wales','Belgium','Germany',
        'Russia','Amsterdam','Scotland','Portugal','Denmark','Czech Republic','Finland','Sweden',
        'Poland', 'Austria','Lithuania','Iceland','Hungary','Ireland']
asia = ['Japan','South Korea','Singapore','Vietnam','Israel','India','Philippines']
north_america = ['U.S.A.','Mexico','Canada','Puerto Rico','Costa Rica','Martinique','Guatemala','St. Lucia','Grenada','Domincan Republic','Honduras', 'Nicaragua']
south_america = ['Ecuador','Peru','Brazil','Venezuela','Colombia','Argentina','Bolivia','Chile','Eucador','Suriname']
africa = ['Sao Tome','South Africa','Madagascar','Ghana']
oceania = ['Fiji','Australia','New Zealand']
country_lists = [europe, asia, north_america, south_america, africa, oceania]
continent_list = ['Europe', 'Asia', 'North_America', 'South_America', 'Africa', 'Oceania']

COUNTRY_CONTINENT_DICT = create_country_continent_dict(country_lists, continent_list)

def parse_company_location(company_location):
  if company_location in COUNTRY_CONTINENT_DICT:
    return COUNTRY_CONTINENT_DICT.get(company_location)
  return 'Unknown'

# Function to parse broad bean origin needs to continent
def parse_broad_bean_origin(broad_bean_origin):
  if broad_bean_origin in COUNTRY_CONTINENT_DICT:
    return COUNTRY_CONTINENT_DICT.get(broad_bean_origin)
  elif ',' in broad_bean_origin:
    return "Mixed"
  return 'Unknown'

# Scaling -- This will probably just be sklearn
# https://scikit-learn.org/stable/modules/preprocessing.html
```

<br>

Formula for MinMaxScaler:
\begin{align}
x_{std} = \frac{x-x_{\min}}{x_{\max}-x_{\min}}
\end{align}
<br>
\begin{align}
x_{scaled} = x_{std} * (\max - \min) + \min
\end{align}

<br>

Whenever we are carrying out a calculation invovling the training data and using it to process the data in some way, such as scaling the data or imputing values it is important to remember we need to do the same operations to the "unseen" data, but as the data represents data that hasn't been gathered, we can't carry out calculations such as working out the mean, what we need to do instead is use the same result we used to process the training (seen) data on the unseen (validation/testing) data.

The main decision I had to make was which kernel I would be using, which as discussed is the function which is used to map lower dimensional data into a higher dimension, I chose to use the default kernel "Radial Basis Function" (RBF) as it is the most generalised kernel, and I don't beleive the regression problem we are tackling is linear. When dealing with RBF the proper choice of "C" and "gamma" is important. "C" is the regularisation parameter, which trades off correct classification against the maximisation of the decisions function margin (Scikit-learn n.d.d), the default "C" value is set to 1, I beleive this will have to be lowered due to the noise in our dataset but 1 is a reasonable starting point, when changing C accuracy on the training may go down but this doesn't show how the model will generalise to unseen data. Gamma is the second important parameter is gamma, which defines how far the the influence of training example reaches, scikit-learns default gamma function is appropriate.

<br>

I chose not to tune parameters whil evaluating against the training data as to avoid overfitting when training and testing on the same data, after researching the parameters, it was also concluded that increase C would reduce error on the training set but could have the inverse affect when generalising to unseen data.

```python
def creat_hash_df(df, column_to_hash, number_of_buckets):
  h = FeatureHasher(n_features=number_of_buckets, input_type='string')
  s = df[column_to_hash]
  f = h.transform(s)
  f = f.toarray()
  return pd.DataFrame(f, columns=[column_to_hash+"_"+str(i) for i in range(1, (number_of_buckets + 1))])
```
